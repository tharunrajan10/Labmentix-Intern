## 3. ***Data Wrangling***

### Data Wrangling Code

```python
# merging the dataset

df = pd.merge(insurance_price_df, personal_details_df, on='Policy no.')
df.head()
```

```python
# Convert categorical variables
df['smoker'] = df['smoker'].astype('category')
df['sex'] = df['sex'].astype('category')
df['region'] = df['region'].astype('category')
```

```python
df.info()
```

```python
# Outlier Detection and Removal
q1 = df['charges in INR'].quantile(0.25)
q3 = df['charges in INR'].quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr
df = df[(df['charges in INR'] >= lower_bound) & (df['charges in INR'] <= upper_bound)]
```

# 8. Insurance Costs for Smokers vs. Non-Smokers by Region  (Donut Chart)

```python
import matplotlib.pyplot as plt

# Summarizing total charges for smokers and non-smokers by region
region_smoker_cost = df.groupby(['region', 'smoker'])['charges in INR'].sum().unstack()

# Creating a subplot for a semi-pie (donut) chart
fig, ax = plt.subplots(figsize=(8, 5), subplot_kw={'aspect': 'equal'})

# Labels and data
labels = []
sizes = []

for region in region_smoker_cost.index:
    labels.append(f"{region} - Smoker")
    sizes.append(region_smoker_cost.loc[region, 'yes'])
    labels.append(f"{region} - Non-Smoker")
    sizes.append(region_smoker_cost.loc[region, 'no'])

# Colors for distinction
colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99', '#c2c2f0', '#ffb3e6', '#c4e17f', '#76d7c4']

# Plotting
wedges, texts, autotexts = ax.pie(
    sizes, labels=labels, autopct='%1.1f%%', startangle=140, colors=colors,
    wedgeprops={'linewidth': 1, 'edgecolor': 'white'}
)

# Creating a donut-style appearance
centre_circle = plt.Circle((0, 0), 0.70, fc='white')
fig.gca().add_artist(centre_circle)

# Adding a title
plt.title("Proportion of Insurance Costs for Smokers vs. Non-Smokers by Region")
plt.show()

```

##### 1. Why did you pick the specific chart?

A donut chart clearly visualizes the proportion of total insurance costs claimed by smokers and non-smokers across different regions.

It provides an easy comparison without excessive details.



##### 2. What is/are the insight(s) found from the chart?

Smokers contribute to a significantly higher share of insurance claims in every region.

Some regions (e.g., Southeast or Northwest) might have a much higher smoker-related claim burden than others.

##### 3. Will the gained insights help creating a positive business impact?
Are there any insights that lead to negative growth? Justify with specific reason.

This insight justifies the need for higher premium rates for smokers.

The company can target specific regions with anti-smoking incentives to reduce claim costs.



If a particular region has a very high percentage of smoking-related claims, it may indicate higher health risks and increase overall policy costs.

Customers from these regions may face premium hikes, potentially making policies less affordable.
