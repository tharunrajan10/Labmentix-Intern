## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.

# Drop unnecessary column
df.drop(df.columns[0], axis=1, inplace=True)
````

````python
# Standardizing column names (removing spaces and converting to lowercase)
df.columns = df.columns.str.replace(" ", "_").str.lower()
````

````python
# Handling missing or incorrect values in 'rating' and 'founded'
df["rating"] = df["rating"].replace(-1, None)  # Replace -1 (missing) with NaN
df["founded"] = df["founded"].replace(-1, None)  # Replace -1 with NaN
````

````python
# Removing non-numeric characters and extracting salary range
df["salary_estimate"] = df["salary_estimate"].str.replace(r"[^\dK-]", "", regex=True)

# Extracting minimum and maximum salary values
df[["salary_min", "salary_max"]] = df["salary_estimate"].str.extract(r"(\d+)K-(\d+)K").astype(float) * 1000
````

````python
# Extracting state from 'location'
df["state"] = df["location"].apply(lambda x: x.split(",")[-1].strip())
````

````python
# Converting 'founded' year to company age
from datetime import datetime
# Get current year dynamically
current_year = datetime.now().year
df["company_age"] = current_year - df["founded"]
````

````python
# Cleaning 'size' column (converting ranges to numeric values)
df["min_size"] = df["size"].str.extract(r"(\d+)").astype(float)
df["max_size"] = df["size"].str.extract(r"to (\d+)").astype(float)

# Setting max_size as NaN for '10000+ employees' or similar cases
df.loc[df["size"].str.contains(r"\+", na=False), "max_size"] = None
````

````python
# Extract minimum and maximum revenue values
df[["revenue_min", "revenue_max"]] = df["revenue"].str.extract(r"(\d+)[^\d]+(\d+)?")

# Convert to float
df["revenue_min"] = df["revenue_min"].astype(float)
df["revenue_max"] = df["revenue_max"].astype(float)

# Convert billion values to millions
billion_mask = df["revenue"].str.contains("billion", case=False, na=False)
df.loc[billion_mask, ["revenue_min", "revenue_max"]] *= 1000  # Convert to millions

# Handling 'Unknown / Non-Applicable' by setting values to NaN
unknown_mask = df["revenue"].str.contains("Unknown / Non-Applicable", case=False, na=False)
df.loc[unknown_mask, ["revenue_min", "revenue_max"]] = np.nan

# Note: All revenue values are now in **millions** of USD
````

````python
# Dropping rows with missing essential values
df.dropna(subset=["salary_min", "salary_max", "rating"], inplace=True)
````

````python
# Reset index
df.reset_index(drop=True, inplace=True)
````

### What all manipulations have you done and insights you found?

1. **Dropped Unnecessary Column:** Removed `Unnamed: 0` since it adds no analytical value.
2. **Standardized Column Names:** Converted all column names to lowercase and replaced spaces with underscores.
3. **Handled Missing/Incorrect Values:**
   - Replaced `-1` values in `rating` and `founded` columns with `NaN`.
4. **Extracted Salary Information:**
   - Extracted minimum and maximum salary estimates from `salary_estimate`.
5. **Extracted State from Location:**  
   - Split `location` column to get the state.
6. **Calculated Company Age:**  
   - Derived `company_age` from `founded` year.
7. **Converted Company Size to Numeric Ranges:**  
   - Extracted min and max employee counts from `size`.
8. **Processed Revenue Information:**  
   - Extracted revenue ranges and converted them to numeric values where applicable.
9. **Dropped Rows with Essential Missing Values:**  
   - Ensured salary and rating information are complete for analysis.
